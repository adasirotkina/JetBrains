```python
cover = pd.read_csv('filtered_cytosines_coverage.tsv', sep = '\t')
cover['chr'] = cover['chrBase'].apply(lambda x: x.split('.')[0])
cover['Base'] = cover['chrBase'].apply(lambda x: int(x.split('.')[1]))
cover.head()
```

```python
cover['chr'].unique()
```

```python
CpG = pd.read_csv('cpgIslandExt.txt', sep = '\t')
#CpG.columns = ['bin', 'chrom', 'chromStart','chromEnd', 'name','length', 'cpgNum', 'gcNum', 'perCpg', 'perGc', 'obsExp']
CpG.head()
```

```python
arr = [0]*len(CpG)

for ch in cover['chr'].unique():
    cover_copy = cover[cover['chr'] == ch]
    CpG_copy = CpG[CpG['chrom'] == ch]
    print(ch, end = ' ')
    for i in CpG_copy.index:
        print(i, end = ' ')
        for j in cover_copy.index:
            if CpG_copy['chromStart'].loc[i] < cover_copy['Base'].loc[j] and CpG_copy['chromEnd'].loc[i] > cover_copy['Base'].loc[j]:
                arr[i] += 1
```
